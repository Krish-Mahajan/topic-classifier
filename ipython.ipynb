## Tokenize Train Data and Creating Vocabularly

```python
train_data,word_index_map,training_tokenized=tokenize_train_data(train_data,3,stopwords) #3 = minfreq 
```

```python
'''
word_index_map = vocabulary
'''
len(word_index_map)
```

```python
#Every word is now indexed from 0 to 39206, which now constitute the Vocab
word_index_map
```

```python
'''
training_tokenization = Contains token list of rows of training_data 
Each document is now availaible as a list of tokens in training_tokenized
'''
training_tokenized[1] 
```

## Vectorizing Training Data

```python
train_data_vector,label_dict = vectorize(train_data,word_index_map,training_tokenized) 
```

```python
'''
train_data_vector = vectorization form of each row of training data
'''
print(len(train_data_vector)) 
print(train_data_vector[111]) 
```

```python
#indexed dictionary of topics
label_dict  
```

```python
'''
Partioning the data , if we dont want to train on full dataset and rather train only on some porportion(fraction) of
the training data and unlabel the remaining labeled data, to be later used by Expectation Maximization Algorithm
Splitting training Data into label and unlabeled as per ",fraction,"probability of looking at each label"
'''

fraction = 0.8 
train_data_vector_label,train_data_vector_unlabel = partition_label_unlabeled(train_data_vector,fraction)
```

```python
'''
Right now , vectors are in form document-term 
Row is the tokenized form of document
Columns are the tokenized term  
Each term is indexed as per the vocabularly  
Now Making term-document matrix of Training label Data and Training unlabel Data
'''
tf_train_label = term_document(train_data_vector_label) 
tf_train_unlabel = term_document(train_data_vector_unlabel)  
```

```python
tf_train_label.shape 
```

```python
'''
document_label: D x L matrix
where document_label(d,l) = 1 if document d belongs to label l 
'''
document_train_label = document_label(train_data_vector_label,tf_train_label) 
```

```python
document_train_label 
```

```python

document_train_label.shape 
```

```python
'''
All Preprocessing done and now time to train the label_data on Naieve Bayes
Calculating 
1) Prior Probabilities of class i,e P(c)
2) Likelihood Probabilities (i,e probabilities of words given topic P(w|c))
'''
nb=snb.NaiveBayes(vocab=word_index_map,label_dict=label_dict)



"Training EM naive Bayes Model..." 
nb.train_semi(tf_train_label,document_train_label,tf_train_unlabel)  
```

```python
print(os.getcwd())
```

```python
"Making distinctive_words for top 10 words in each topic"
nb.top_10_words('./com/output/distinctive_words.txt')

```

```python
"Saving Trained Model.."
f = open("./com/output/model-file.p","w")
pickle.dump(nb,f)
f.close()

```

# Testing Data

```python
print("opening Saved semi supervised naieve bayes model") 
ff = open("./com/output/model-file.p","r")
nb=pickle.load(ff)
ff.close() 
```

```python
"Reading Testing Data..."
test_data=read_data('./data/testResult')   
```

```python
"Tokinizing Test Data"
testing_tokenized =tokenize_test_data(test_data,nb.vocab,3,stopwords)    
```

```python
"Vectorizing Test Data..."
word_index_map=nb.vocab
label_dict = nb.label_dict
test_data_vector,label_dict= vectorize(test_data,word_index_map,testing_tokenized,label_dict)
```

```python
"Making term document matrix of testing Data"
tf_test = term_document(test_data_vector) 
```

```python
"Making  document_label_test on testing label"
document_label_test = document_label(test_data_vector,tf_test) 
```

```python
"Predicting test Data"
predict=nb.predict_proba_all(tf_test)
"Accuracy is",nb.accuracy(predict,document_label_test)
```
